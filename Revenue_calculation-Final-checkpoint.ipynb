```python
import pandas as pd
```

```python
rank_data = pd.read_excel("Data_Set_for_Data_Analyst_Exercise.xlsx", sheet_name="Rank Data")
rank_data[:5]
```

```python
business_data = pd.read_excel("Data_Set_for_Data_Analyst_Exercise.xlsx", sheet_name="Business Data")
business_data[:5]
```

```python
other_data = pd.read_excel("Data_Set_for_Data_Analyst_Exercise.xlsx", sheet_name="Other Data")
other_data = other_data.dropna(how="all")
other_data = other_data.dropna(how="all", axis=1).reset_index(drop=True)
conversion_rate = other_data.loc[0][1]
other_data.columns = ["Rank", "CTR"]
other_data = other_data.drop(other_data.index[[0,1]]).reset_index(drop=True)
other_data[:5]
```

```python
merged_data = rank_data.merge(business_data, on = "Category", how="left")
merged_data = merged_data.merge(other_data, on = "Rank", how="left")
merged_data[:5]
```

```python
#Calculating revenue
```

```python
merged_data = merged_data.dropna() #Dropping rows with no CTR value
merged_data["Estimated_revenue"] = merged_data.apply(lambda row: row["MSV"]*row["CTR"]*conversion_rate*row["AVG Order Value, Organic Channel (Monthly)"], axis=1)
merged_data[:5]
```

```python
#Total revenue
import numpy as np
total_revenue = np.sum(merged_data.Estimated_revenue)
print( "Total revenue is:" + str(total_revenue))
```

```python
#Top 5 Keywords with revenue
merged_data = merged_data.sort_values("Estimated_revenue", ascending=False)
merged_data[["Keyword", "Estimated_revenue"]][:5]
```

```python
#Bottom 5 Keywords with revenue
merged_data = merged_data.sort_values("Estimated_revenue", ascending=False)
merged_data[["Keyword", "Estimated_revenue"]][-5:]
```

```python
#Top 5 Category with revenue
grouped_merged_data = merged_data.groupby("Category")[["Estimated_revenue"]].sum()
grouped_merged_data = grouped_merged_data.sort_values("Estimated_revenue", ascending=False)
grouped_merged_data = grouped_merged_data.reset_index()
grouped_merged_data[:5]
```

```python
#Bottom 5 Category with revenue
grouped_merged_data = merged_data.groupby("Category")[["Estimated_revenue"]].sum()
grouped_merged_data = grouped_merged_data.sort_values("Estimated_revenue", ascending=False)
grouped_merged_data = grouped_merged_data.reset_index()
grouped_merged_data[-5:]
```
